Dada uma sequência nucleotídica pelo usário, verificar se sequência corresponde a um fragmento de DNA ou um gene completo eucarioto. Isto é, verifique se sequência inicia com ATG e termina em stop codon: TAG, TAA, TGA, além da sequência ser divisivel por 3.

```python
# seu código aqui
# seq1 de exemplo: 'ATCGATCGTAGCTAGCTAGCTATCAGTAGCTCTATA'
# seq2 de exemplo: 'ATGGATCGTAGCTAGCTAGCTATCAGTAGCTCTTAA'
# seq3 de exemplo: 'ATGGATCGTAGCTAGCTAGCTATCAGTGCTCTTAA'
seq = input('Digite sequência: ')
# este strip tira o pula-linha ('\n')que tem no final
seq = seq.strip() 
# lembrando que:
# ATG: Start Codon
# TAG', 'TAA', 'TGA': Stop codon
# primeuira condiçao start codon na posicao 0
# segunda condição stop na ultima posilçao
# terceira cond. ser div por 3
stop_codon = ['TAG', 'TAA', 'TGA']
if (seq.find('ATG') == 0 and (seq[len(seq)-3:len(seq)] == stop_codon[0] or stop_codon[1] or stop_codon[2])) and (len(seq)/3) == 1:
  print('Si es')
else:
  print('No es')
```

```python
test = 'ATCGATCGTAGCTAGCTAGCTATCAGTAGCTCTATA'
```

### Solução

```python
# seq1 de exemplo: 'ATCGATCGTAGCTAGCTAGCTATCAGTAGCTCTATA'
# seq2 de exemplo: 'ATGGATCGTAGCTAGCTAGCTATCAGTAGCTCTTAA'
# seq3 de exemplo: 'ATGGATCGTAGCTAGCTAGCTATCAGTGCTCTTAA'
seq = input('Digite sequência: ')
seq = seq.strip() # este strip tira o pula-linha ('\n')que tem no final
ini_codon = 'ATG'
stop_codon = ['TAG', 'TAA', 'TGA']
seq_ini = seq[0:3]
seq_end = seq[len(seq)-3:len(seq)]

if (seq_ini == ini_codon) and (seq_end in stop_codon) and (len(seq)%3 == 0):
  print('Gene eucarioto')
else:
  print('Fragmento DNA')
```

```python
# seq de exemplo: 'ATCGATCGTAGCTAGCTAGCTATCAGTAGCTCTATA'
seq = input('Digite sequência: ')
# este strip tira o pula-linha ('\n')que tem no final
seq = seq.strip() 
# A, T, C, G
counts = [0,0,0,0]
for i in seq:
  if(i =="A"):
    counts[0] +=  1
  elif(i =="T"):
    counts[1] +=  1
  elif(i =="C"):
    counts[2] +=  1
  elif(i =="G"):
    counts[3] +=  1
freq_A = counts[0]/len(seq)
freq_T = counts[1]/len(seq)
freq_C = counts[2]/len(seq)
freq_G = counts[3]/len(seq)
print(f'A frequência de resíduo de adenina é {round(freq_A,3)}')
print(f'A frequência de resíduo de Timina é {round(freq_T,3)}')
print(f'A frequência de resíduo de Citosina é {round(freq_C,3)}')
print(f'A frequência de resíduo de Guanina é {round(freq_G, 3)}')
```

## Exercício 5
Escrever uma função que recebe uma sequência nucleotídica, calcular a frequência de resíduos de cada base.

Sem utilizar nenhuma função da classe str

(Usar for e if)

Usando Dicionário

### Solução

```python
def calc_freq(seq_file):
  # dicionário para contar - base é a "chave" (key) e sua contagem é o "valor"(value)
  dic = {"A": 0, "T": 0, "C": 0, "G":0}
  with open(seq_file, 'r', encoding='utf-8-sig') as f:
      for line in f:
        print(line)
        seq = line.strip() 
        for i in seq:
          dic[i]+=1
  freq_A = dic["A"]/len(seq)
  freq_T = dic["T"]/len(seq)
  freq_C = dic["C"]/len(seq)
  freq_G = dic["G"]/len(seq)
  print(f'A frequência de resíduo de adenina é {round(freq_A,3)}')
  print(f'A frequência de resíduo de Timina é {round(freq_T,3)}')
  print(f'A frequência de resíduo de Citosina é {round(freq_C,3)}')
  print(f'A frequência de resíduo de Guanina é {round(freq_G, 3)}')
```

```python
# input: filename of xls dataset (gene expression)
# returns: pandas dataframe 
def read_dataset(filename):
  df = pd.read_excel(filename, index_col=0)
  if 'Gene_Symbol' in list(df.columns):
    df = df.drop(columns=['Gene_Symbol'])
  columns = list(df.columns)
  df_data = pd.DataFrame (df, columns=columns, index=df.index)
  return df_data
```

```python
df_ex = read_dataset("Broadbendt2015.xlsx")
df_ex
```

```python
# Given a dataset (gene expression) applies log2 and normalize (around time) mean = 0 std = 1
# input: filename of xls dataset (gene expression)
# returns: pandas dataframe (with log2 normalized)
def get_df_data_log_norm(filename):
  df_data = read_dataset(filename)
  np_data = np.log(df_data.to_numpy()+1) / np.log(2)
  columns = list(df_data.columns)
  df_data_log = pd.DataFrame (np_data, columns=columns, index=df_data.index)
  df_data_log_norm = (df_data_log.T - df_data_log.mean(axis=1))/ df_data_log.std(axis=1)
  df_data_log_norm = df_data_log_norm.T
  df_data_log_norm=df_data_log_norm.dropna()

  return df_data_log_norm
```

```python
# getting gene expression dataset (already transformed log2 and normalized)
df_data_log_norm = get_df_data_log_norm("Broadbendt2015.xlsx")

df_data_log_norm
```

```python
# Given a dataset (gene expression df), a list of genes or a unique gene, stage of plasmodium(optional): returns gene expression (in this <stage> or in each time point)
# input: gene expression df (with log2 and mean 0), list of genes or unique gene, name (string) of the plasmodium stage { "ring": "6h":"20h", "trophozoite": "24h":"36h", "schizont": "36h":"48h"}
# returns: pandas dataframe (with log2 and mean 0) of <l> in <stage>
def get_genes_expression(data,l, stage=False):
  if(not stage):
    return data.loc[ l , : ]
  if(stage == "ring"):
    return data.loc[ l , "6h":"20h" ]
  if(stage == "trophozoite"):
    return data.loc[ l , "24h":"36h" ]
  if(stage == "schizont"):
    return data.loc[ l , "36h":"48h" ]
```
